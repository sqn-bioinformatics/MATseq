```python
def raw_to_zscore(DataFrame):
    outputDF = pd.DataFrame(
        zscore(DataFrame, axis=0), index=DataFrame.index, columns=DataFrame.columns
    )
    return outputDF


def raw_to_rpm(DataFrame):
    outputDF = pd.DataFrame()
    for sample in DataFrame.columns:
        total_reads = DataFrame[sample].sum()
        outputDF[sample] = DataFrame[sample].apply(
            lambda x: (x / total_reads) * 1000000
        )
    return outputDF


def get_classes(lista):
    output = []
    for i in lista:
        output.append(i.split("_")[2])
    return output
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


# Split your data into 18 smaller dataframes
num_samples = 12  # Replace with the actual number of samples
dataframes = []
data1 = data.copy()

# Split the data into equal parts
split_data = np.array_split(data1, num_samples)


for i, split in enumerate(split_data):
    dataframes.append(pd.DataFrame(split))
```

```python
len(dataframes)
```

```python
# Create a 2x2 grid for plotting
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Flatten the 2D array of axes to make it easier to work with
axes = axes.ravel()


# Loop through your dataframes and plot each one
for i, data in enumerate(dataframes[8:12]):  # Assuming you have 18 dataframes
    ax = sns.heatmap(data, cmap="YlGnBu", ax=axes[i])
    ax.set_xlabel("Reference Pyrogen", fontsize=15)
    ax.set_ylabel("Sample", fontsize=15, labelpad=10)
    ax.set_title(f"Prediction MLPC after fastp", fontsize=15)

# If you have fewer than 4 dataframes, you can add empty subplots to fill the grid
for i in range(len(dataframes), 4):
    fig.delaxes(axes[i])

# Adjust layout
plt.tight_layout()

# Show or save the plot
plt.show()  # Use plt.savefig("your_filename.png") to save the plot to a file
```
